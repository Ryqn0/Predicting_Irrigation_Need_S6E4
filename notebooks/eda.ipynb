```python
train_df = pd.read_csv("../data/raw/train.csv")
```

```python
from sklearn.model_selection import train_test_split

train, val = train_test_split(
    train_df, test_size=0.2, stratify=train_df["Irrigation_Need"], random_state=42
)

X_train, train_enc = build_final_features(train)
y_train = X_train["Irrigation_Need"]
X_train = X_train.drop(columns=["id", "Irrigation_Need"])
X_val, _ = build_final_features(val, encodings=train_enc)
y_val = X_val["Irrigation_Need"]
X_val = X_val.drop(columns=["id", "Irrigation_Need"])
```

```python
import lightgbm as lgb 

params = {
    "objective":         "multiclass",
    "num_class":         3,
    "metric":            "multi_logloss",
    "n_estimators":      10000,

    # tree structure
    "num_leaves":        127,
    "max_depth":         -1,
    "min_child_samples": 100,    # higher = less overfit on 630k rows

    # stochastic regularization
    "feature_fraction":  0.8,
    "bagging_fraction":  0.8,
    "bagging_freq":      5,

    # L1/L2
    "lambda_l1":         0.1,
    "lambda_l2":         1.0,

    "learning_rate":     0.05,
    "n_jobs":            -1,
    "seed":              42,
    "verbose":           -1,
}

model = lgb.LGBMClassifier(**params)

model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    callbacks=[
        lgb.early_stopping(stopping_rounds=50, verbose=True),
        lgb.log_evaluation(period=200),
    ],
)

print(f"Best iteration: {model.best_iteration_}")
```

```python
from sklearn.metrics import balanced_accuracy_score

y_val_pred = model.predict(X_val)
balanced_acc = balanced_accuracy_score(y_val, y_val_pred)
print(f"Balanced Accuracy: {balanced_acc:.4f}")
```

```python
import xgboost as xgb

xgb_params = {
    "objective":        "multi:softprob",
    "num_class":        3,
    "eval_metric":      "mlogloss",
    "n_estimators":     10_000,

    # tree structure (level-wise, so max_depth is stricter than LGBM's num_leaves)
    "max_depth":        6,
    "min_child_weight": 10,      # equivalent to min_child_samples

    # stochastic regularization
    "subsample":        0.8,
    "colsample_bytree": 0.8,
    "colsample_bylevel": 0.8,

    # L1/L2
    "reg_alpha":        0.1,
    "reg_lambda":       1.0,

    "learning_rate":    0.05,
    "tree_method":      "hist",  # fast on large datasets, required for GPU too
    "n_jobs":           -1,
    "seed":             42,

    "early_stopping_rounds": 50,
}

xgb_model = xgb.XGBClassifier(**xgb_params)

xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=200,
)

print(f"Best iteration: {xgb_model.best_iteration}")

y_val_pred_xgb = xgb_model.predict(X_val)
balanced_acc_xgb = balanced_accuracy_score(y_val, y_val_pred_xgb)
print(f"Balanced Accuracy: {balanced_acc_xgb:.4f}")
```

```python
from catboost import CatBoostClassifier, Pool

train_pool = Pool(X_train, y_train)
val_pool   = Pool(X_val,   y_val)


cb_params = {
    "loss_function":      "MultiClass",
    "eval_metric":        "MultiClass",
    "iterations":         10_000,
    "early_stopping_rounds": 50,

    # tree structure
    "depth":              6,
    "min_data_in_leaf":   100,

    # regularization
    "l2_leaf_reg":        3.0,
    "random_strength":    1.0,
    "bagging_temperature": 1.0,

    "learning_rate":      0.05,
    "border_count":       128,   # bins for numerical features
    "verbose":            200,
    "random_seed":        42,
}

cb_model = CatBoostClassifier(**cb_params)

cb_model.fit(train_pool, eval_set=val_pool)

print(f"Best iteration: {cb_model.best_iteration_}")

y_val_pred_cb = cb_model.predict(X_val)
balanced_acc_cb = balanced_accuracy_score(y_val, y_val_pred_cb)
print(f"Balanced Accuracy: {balanced_acc_cb:.4f}")
```
